## Producer: read new data and send to comsumer 1 

```python
from kafka import KafkaProducer
import time
import os

# Set up the Kafka producer
producer = KafkaProducer(
    bootstrap_servers='localhost:9092',
    max_request_size=8 * 1024 * 1024  # Set as 8MB
)

def send_new_files_to_kafka(directory):
    """ Continuously monitor a directory for new files and send to Kafka """
    seen_files = set()
    while True:
        current_files = set(os.listdir(directory))
        new_files = current_files - seen_files
        for file in new_files:
            if file.endswith(".wav"):
                file_path = os.path.join(directory, file)
                # Read the actual binary content of the .wav file
                with open(file_path, "rb") as f:
                    file_content = f.read()
                print(f"File size: {len(file_content)} bytes")
                try:
                    # Send the binary content to Kafka topic 'audio_files' and wait for it to finish
                    future = producer.send('audio_files', value=file_content)
                    result = future.get(timeout=10)  # Wait for send to complete
                    print(f"Sent content of {file_path} to Kafka, offset: {result.offset}")
                except Exception as e:
                    print(f"Failed to send {file_path} to Kafka: {e}")
        seen_files = current_files
        time.sleep(5)  # Check every 5 seconds

# Specify directory to watch
audio_directory = "./Datasets/New"
send_new_files_to_kafka(audio_directory)
```
